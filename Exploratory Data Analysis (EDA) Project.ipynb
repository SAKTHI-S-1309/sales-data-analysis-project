# Exploratory Data Analysis (EDA) Project

## Problem Statement
The objective of this project is to analyze retail sales data to identify patterns, trends, and key factors influencing sales across categories, regions, and customer segments.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.columns
```

```python
df.info()
df.describe()
df.isnull().sum()
```

## Data Cleaning

```python
df.isnull().sum()
df.drop_duplicates(inplace=True)

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
df[['Order Date', 'Year', 'Month']].head()
```

## Category-wise Sales Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.show()
```

### Observation:
The Technology category shows the highest sales, while Furniture has relatively lower performance.

## Region-wise Sales Analysis

```python
sns.barplot(x='Region', y='Sales', data=df)
plt.title("Sales by Region")
plt.show()
```

### Observation:
The West region generates the highest sales, whereas the South region shows lower sales performance.

## Segment-wise Sales Analysis

```python
sns.barplot(x='Segment', y='Sales', data=df)
plt.title("Sales by Segment")
plt.show()
```

### Observation:
The Consumer segment contributes the largest share of total sales compared to other segments.

## Month-wise Sales Analysis

```python
monthly_sales = df.groupby('Month')['Sales'].sum().sort_index()

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.show()
```

### Observation:
Sales fluctuate across months, with noticeable peaks indicating seasonal demand patterns.

## Sub-Category-wise Sales Analysis

```python
sns.barplot(x='Sub-Category', y='Sales', data=df)
plt.xticks(rotation=45)
plt.title("Sales by Sub-Category")
plt.show()
```

### Observation:
Certain sub-categories contribute significantly to total sales, dominating others in performance.

## Key Insights

- The Technology category generates the highest sales, indicating strong market demand.

- The West region shows the highest sales performance compared to other regions.

- The Consumer segment contributes the largest share of total sales.

- Certain sub-categories contribute significantly to overall revenue.

- Sales vary across months, indicating seasonal trends.

- A small number of products dominate total sales contribution.

- Some regions and categories underperform, suggesting scope for improvement.
